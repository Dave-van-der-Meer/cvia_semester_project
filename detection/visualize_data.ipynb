### We show a number of random images from the dataset.

```python
rows = 3
cols = 4

fig, axes = plt.subplots(rows, cols, figsize=(15, 15))

for i in range(rows):
    for j in range(cols):
        dataset.visualize(randint(0, 2000),size = (10,10),ax=axes[i][j])
        axes[i][j].axis('off')
fig.tight_layout() 
```
